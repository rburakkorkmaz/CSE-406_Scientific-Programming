```python
def random_number_generator(n, a, b, filename):
  # Let's use list comprehension with randrange function
  random_numbers = [rnd.randrange(a, b, 1) for _ in range(0, n)]
  
  with open(filename + ".txt", "w") as file:
    for number in random_numbers:
      file.write(str(number) + " ")


random_number_generator(10, 5, 100, "deneme")
```

  <li>(j) Add any more functionality that you think is useful but I failed to mention.</li>

```python
# Answer to (a)
class Rational:
  p = None
  q = None
  def __init__(self, p, q) -> None:
    if q == 0:
        raise ValueError('Denominator must not be zero')
    if not isinstance(p, int):
        raise TypeError('Numerator must be an integer')
    if not isinstance(q, int):
        raise TypeError('Denominator must be an integer')

    # Answer to (c)
    common_divisor = math.gcd(p, q)

    self.p = p // common_divisor
    self.q = q // common_divisor


  # Answer to (d)
  def __repr__(self):
      return f"{self.p}/{self.q}"

  def __str__(self):
      return f"{self.p}/{self.q}"

  # Answer to (e)
  def __add__(self, r2):
    return Rational(self.p * r2.q + r2.p * self.q, self.q * r2.q)
  
  # Answer to (f)
  def __sub__(self, r2):
    return Rational(self.p * r2.q - r2.p * self.q, self.q * r2.q)
  
  # Answer to (g)
  def __mul__(self, r2):
    return Rational(self.p * r2.p, self.q * r2.q)
  
  # Answer to (h)
  def __truediv__(self, r2):
    return Rational(self.p * r2.q, self.q * r2.p)

  # Answer to (i)
  def __eq__(self, r2):
    return self.p == r2.p and self.q == r2.q

  # Answer to (j)
  def __float__(self):
    return float(self.p) / float(self.q)
```

<h1><strong>Exercise 8.4:</strong> Sparse and dense vectors</h1>
In exercise 6.5 you implemented functions for sparse and dense vector multiplications using lists and dictionaries. However, this is a bit clumsy to use in practice. Really, we would like to represent sparse and dense vectors as classes, this way we can overload operators such as + ( add ) and get sensible
output. For example, using + on two dense vectors implemented as lists would append the second vector to the first, instead of adding the two together.

Implement sparse and dense vectors. Both classes should have the following capabilities:

<ul>
  <li>(a) Print vector</li>
  <li>(b) Add two vectors (both if other is dense and sparse)</li>
  <li>(c) Multiply two vectors (both if other is dense and sparse)</li>
</ul>

Do re-use your code from the previous exercise.
Hint: isinstance() might be useful.

```python
class DenseVector:
  vector = {}
  def __init__(self, vector):
      self.vector = vector
  
  # Print function override
  def __repr__(self):
      return f"{self.vector}" 
  
  # Summation (+) operator function override
  def __add__(self, v2):
    temp = {}

    # Checking if other vector is dense or sparse
    if isinstance(v2, DenseVector):
      for (key, value) in self.vector.items():
        temp[key] = value + v2.vector[key]
    elif isinstance(v2, SparseVector):
      for index, (key, value) in enumerate(self.vector.items()):
        temp[key] = value + v2.vector[index]

    return DenseVector(temp)

  # Multiplication (*) operator function override
  def __mul__(self, v2):
    temp = {}

    # Checking if other vector is dense or sparse
    if isinstance(v2, DenseVector):
      for key, value in self.vector.items():
        temp[key] = value * v2.vector[key]
    elif isinstance(v2, SparseVector):
      for index, (key, value) in enumerate(self.vector.items()):
        temp[key] = value * v2.vector[index]

    return DenseVector(temp)
```

```python
class SparseVector:
  vector = []
  def __init__(self, vector):
      self.vector = vector
  
  # Print function override
  def __repr__(self):
      return f"{self.vector}" 
  
  # Summation (+) operator function override
  def __add__(self, v2):
    temp = []
    for index, value in enumerate(self.vector):
      temp.append(value + v2.vector[index])

    return SparseVector(temp)

  # Multiplication (*) operator function override
  def __mul__(self, v2):
    temp = []
    for index, value in enumerate(self.vector):
      temp.append(value * v2.vector[index])

    return SparseVector(temp)
```

Write a class mySet that has the same basic functionality as the Python set data structure. Base your implementation on a dictionary.

```python
# Set class 
class MySet:
  my_set = {}
  def __init__(self):
      pass

  # Print function override
  def __str__(self):
    set_string = "{"
    for index, (key, value) in enumerate(self.my_set.items()):
      if index + 1 != len(self.my_set):
        set_string += f"{key}, "
      else:
        set_string += f"{key}" + "}"
    return set_string

  # Insert function to insert values
  def insert(self, value):
    self.my_set[value] = True

    # Sort every time insert some value
    self.sort_my_set()

  # Remove function to remove values
  def remove(self, value):
    self.my_set.pop(value, None)

    # Sort every time remove some value
    self.sort_my_set()

  # Sort function to manage values of my_set
  def sort_my_set(self):
    self.my_set = {element:True for element in sorted(self.my_set, key=str)}
```

In this exercise, we will implement a binary search tree. See http://en.wikipedia.org/wiki/Binary_search_tree for an explanation.
<ul>
  <li>(a) Define a class Node, and write the constructor, which takes one argument, value, and initializes
the left and right children to None.</li>
  <li>(b) Write a function to print the tree.</li>
  <li>(c) Write a function that inserts a new value in the tree at the right location.</li>
  <li>(d) Write a function that looks up a value in the tree.</li>
  <li>(e) Write a function that removes a value from the tree.</li>
</ul>

```python
# ANSWER (A)
class Node:
  l_child = None
  r_child = None
  parent = None
  value = None
  def __init__(self, value):
    self.value = value

class BinarySearchTree:
  def __init__(self, root_value):
    self.root = Node(root_value)

  # ANSWER (B)
  def inorder_print(self, n):
    """ Inorder print function. It uses recursion to print Binary Search Tree """
    if n != None:
      self.inorder_print(n.l_child)
      print(n.value, end=" ")
      self.inorder_print(n.r_child)

  def print_tree(self):
    """ This method is only clean coe purpuse. I did not wanted to 
    insert root as parameter to function. """
    print("Inorder")
    self.inorder_print(self.root)
    print("\n")

  
  def find_value(self, value):
    """
      Finds value in the tree, return (True, Node) tuple. 
      If no such value in the tree return (False, None) tuple
    """
    # Create a temporary variable to travel between nodes
    # Start from root
    indicator = self.root

    while True:
      # If we actually find the value, we return True, Node
      if value == indicator.value:
        return True, indicator
      
      # If value we searching for is bigger than current node we are sitting on
      # then move cursor to right child of current node
      elif value > indicator.value:
        if indicator.r_child != None:
          indicator = indicator.r_child
        # If there is no such child in the right of cursor. Then value does not 
        # exist in the tree
        else:
          return False, None
      
      # If value we searching for is less than current node we are sitting on
      # then move cursor to left child of current node
      elif value < indicator.value:
        if indicator.l_child != None:
          indicator = indicator.l_child
        # If there is no such child in the left of cursor. Then value does not 
        # exist in the tree
        else:
          return False, None

  def insert(self, value):
    """
      This function inserts a new value to tree.
    """
    indicator = self.root

    while value >= indicator.value:
      if indicator.r_child != None:
        indicator = indicator.r_child
      else:
        indicator.r_child = Node(value)
        indicator.r_child.parent = indicator
        break
    
    while value < indicator.value:
      if indicator.l_child != None:
        indicator = indicator.l_child
      else:
        indicator.l_child = Node(value)
        indicator.l_child.parent = indicator
        break
```

```python
bst = BinarySearchTree(5)

# ANSWER (C)
bst.insert(6)
bst.insert(1)
bst.insert(32)
bst.insert(53)
bst.insert(22)
bst.insert(9)

# ANSWER (B)
bst.print_tree()

# ANSWER (D)
is_found, n = bst.find_value(32)
print(f"Is number 32 in the tree: {is_found}")
print(f"If found what are the relative nodes of number 32 node: Left Child: {n.l_child.value}, Right Child: {n.r_child.value}")
is_found, n = bst.find_value(99)
print(f"Is number 99 in the tree: {is_found}")
```
